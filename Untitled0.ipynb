```python
import pandas as pd

# Φόρτωση του dataset από τον σύνδεσμο
df = pd.read_csv('https://storage.googleapis.com/courses_data/Assignment%20CSV/finance_liquor_sales.csv')
```

```python
# Επισκόπηση των πρώτων γραμμών
print(df.head())

# Στοιχεία για τους τύπους δεδομένων και τις κενές τιμές
print(df.info())
print(df.describe())
```

```python
# Αφαίρεση γραμμών με κενά δεδομένα
df = df.dropna()
```

```python
# Μετατροπή της στήλης ημερομηνίας σε datetime format
df['date'] = pd.to_datetime(df['date'])

# Φιλτράρισμα για το χρονικό διάστημα 2016-2019
df = df[(df['date'].dt.year >= 2016) & (df['date'].dt.year <= 2019)]
```

```python
# Εύρεση του πιο δημοφιλούς προϊόντος ανά ταχυδρομικό κώδικα
popular_items = df.groupby(['zipcode', 'item_description']).size().reset_index(name='count')
popular_items = popular_items.loc[popular_items.groupby('zipcode')['count'].idxmax()]
```

```python
# Υπολογισμός συνολικών πωλήσεων για το χρονικό διάστημα
total_sales = df['sale_dollars'].sum()

# Υπολογισμός πωλήσεων ανά κατάστημα και ποσοστού
store_sales = df.groupby('store_number')['sale_dollars'].sum().reset_index()
store_sales['sales_percentage'] = (store_sales['sale_dollars'] / total_sales) * 100
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Οπτικοποίηση του ποσοστού πωλήσεων ανά κατάστημα
plt.figure(figsize=(12, 8))
sns.barplot(data=store_sales, x='store_number', y='sales_percentage')
plt.title('Ποσοστό Πωλήσεων ανά Κατάστημα (2016-2019)')
plt.xlabel('Αριθμός Καταστήματος')
plt.ylabel('Ποσοστό Πωλήσεων (%)')
plt.xticks(rotation=90)
plt.show()
```
